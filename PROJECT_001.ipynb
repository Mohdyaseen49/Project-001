```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
yz = pd.read_csv("/content/drive/MyDrive/StudentPerformanceFactors.csv")
```

```python
yz.head()
```

```python
yz.isnull()
```

```python
yz.isnull().sum()
```

```python
mode_value1 = yz['Teacher_Quality'].mode()
print(mode_value1)
```

```python
mode_value2 = yz['Parental_Education_Level'].mode()
print(mode_value2)
```

```python
mode_value3 = yz['Distance_from_Home'].mode()
print(mode_value3)
```

```python
yz['Teacher_Quality'].fillna(yz['Teacher_Quality'].mode()[0], inplace=True)
yz['Parental_Education_Level'].fillna(yz['Parental_Education_Level'].mode()[0], inplace=True) # Changed 'Parental_Education_level' to 'Parental_Education_Level'
yz['Distance_from_Home'].fillna(yz['Distance_from_Home'].mode()[0], inplace=True)
```

```python
yz.isnull().sum()
```

```python
#How does the number of hours studied per week correlate with the final exam score?
correlation = yz['Hours_Studied'].corr(yz['Exam_Score'])
print(f"The correlation between Hours Studied Per Week and Final Exam Score is: {correlation}")
```

```python
#for visual representation
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=yz, color='cyan', alpha=0.9)
plt.title(f'Scatter Plot: Hours Studied vs Exam Score\nCorrelation: {correlation:.2f}')
plt.xlabel('Hours Studied per Week')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
#Do private school students perform better in exams than public school students?
avg_scores = yz.groupby('School_Type')['Exam_Score'].mean()

print(f"Average Exam Score for Private School Students: {avg_scores.get('Private', 'N/A')}")
print(f"Average Exam Score for Public School Students: {avg_scores.get('Public', 'N/A')}")

if avg_scores.get('Private', 0) > avg_scores.get('Public', 0):
    print("Private school students tend to perform better in exams than public school students.")
elif avg_scores.get('Private', 0) < avg_scores.get('Public', 0):
    print("Public school students tend to perform better in exams than private school students.")
else:
    print("Private and public school students perform equally in exams.")
```

```python
#for visual repersentation
plt.figure(figsize=(8, 6))
sns.barplot(x='School_Type', y='Exam_Score', data=yz, palette='mako')
plt.title('Average Exam Score by School Type')
plt.xlabel('School Type')
plt.ylabel('Average Exam Score')
plt.grid(axis='y')

```
